```python
f_list = ['phone','tv','ac']
s_list = []

def popping(f_list,s_list):
    while f_list:
        pop_out1 = f_list.pop()
        print("popping out: " + pop_out1)
        s_list.append(pop_out1)
def after_pop(s_list):
    print('\nItems in list2 are: ')
    for every in s_list:
        print(every)
popping(f_list,s_list)
after_pop(s_list)
```

```python
# lest test it
lista = ['banana','apple','mango']
listb = []
while lista:
    took = lista.pop()
    listb.append(took)
for eachone in listb:
    print(eachone)
```

```python
# copying: only works with functions
listc = ['banana','apple','mango']
listd = []
def new(listc,listd):
    while listc[:]:  # applied copy method here[:]
        took = listc.pop()
        listd.append(took)
def new2(listd):
    for eachone in listd:
        print(eachone)
new(listc[:],listd) # applied copy method here also
new2(listd)
```

```python
class Battery:
    def __init__(self,battery_per = 60):
        self.battery_per = battery_per
    def describe_battery(self):
        print('this car has ' + str(self.battery_per )+ '-kwh Battery' )
    def get_range(self):
        if self.battery_per == 60:
            range = 140
        elif self.battery_per == 80:
            range = 180
            print('already full')
        print('this car can go ' + str(range) + 'miles on a full charge')
    def upgrade_battery(self):
        if self.battery_per == 60:
            self.battery_per = 80
            print('Upgraded the Battery to 80kwh')
        else:
            print('Battery is already Upgraded')        
```

```python
class ElectricCar(Car):
    def __init__(self,name,model,year):
        super().__init__(name,model,year)
        self.battery = Battery()
```

```python
ecar = ElectricCar('Tesla','S',2019)
print(ecar.description()+ '\n')
print('Checking the Electric car Battery: \n')
ecar.battery.describe_battery()
ecar.battery.upgrade_battery()   # upgrading the battery
ecar.battery.upgrade_battery()   # upgrading again
```

```python
# json.dump() --> to store data

import json

numbers = [1,2,3,4,5,6,7,8]

filename = 'numbers.json'

with open(filename, 'w') as f:
    json.dump(numbers, f)
```

```python
# json.load() --> loading the file

import json

with open('numbers.json') as file_object:
    show = json.load(file_object)
print(show)
```

```python
# to store multiple languages in response:

import import_ipynb
import unittest
from survey import AnonymousSurvey

class TestAnonymousSurvey(unittest.TestCase):
    def test_store_response(self):
        question = 'Your Favourite Programming Language '
        my_survey = AnonymousSurvey(question)
        my_survey.store_response("Python")
        self.assertIn('Python',my_survey.responses)
        
    def test_store_three_responses(self):
        question = 'Your Favourite Prigramming Language '
        my_survey = AnonymousSurvey(question)
        # my_survey.store_response('python') --> as it can handle only one response so we are using the following code,
        responses = ['Python', 'Ruby', 'java']
        for each in responses:
            my_survey.store_response(each)
        for every in responses:
            self.assertIn(every, my_survey.responses)
            
if __name__ == '__main__':
    unittest.main(argv = ['first-arg-is-ignored'], exit = False)
```

```python
# The setUp() Method
```

```python
#  a setUp() method that allows you to create these objects once and then use them in each of your test methods.
# When you include a setUp() method in a TestCase class, Python runs the setUp() method before running each method...
# ...starting with test_. Any objects created in the setUp() method are then available in each test method you write. 
```

```python
import import_ipynb
import unittest
from survey import AnonymousSurvey
```

```python
class TestAnonymousSurvey(unittest.TestCase):
    def setUp(self):
        question = 'Your favourite Programming Languages'
        self.my_survey = AnonymousSurvey(question)
        self.responses = ['Python','Ruby','C#']
    
    def test_store_single_response(self):
        self.my_survey.store_response(self.responses[0])
        self.assertIn(self.responses[0], self.my_survey.responses)
        
    def test_store_three_responses(self):
        for each1 in self.responses:
            self.my_survey.store_response(each1)
        for every1 in self.responses:
            self.assertIn(every1, self.my_survey.responses)
            
if __name__ == '__main__':
    unittest.main(argv = ['first-arg-is-ignored'], exit = False)
```
